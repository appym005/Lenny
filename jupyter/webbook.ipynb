```python



def clean(s):
    c = ''
    s = s[18:]
    s.strip()
    for ch in s:
        if((ch>='a' and ch<= 'z') or (ch>='A' and ch<='Z') or ch == ' ' or ch == ',' or ch == '.'):
            c += ch
        else:
            c += ' '
    return c
```
